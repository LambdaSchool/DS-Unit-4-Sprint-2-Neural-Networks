### NAND Gate Prediction (Perceptron Class)

```python
class Perceptron(object):
    
    def __init__(self, rate = 0.01, niter = 1000):
        self.rate = rate
        self.niter = niter

    # Sigmoid activation function and its derivative
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        return self.sigmoid(x) * (1 - self.sigmoid(x))

    def fit(self, X, y):
        """Fit training data
        X : Training vectors, X.shape : [#samples, #features]
        y : Target values, y.shape : [#samples]
        """

        # Randomly Initialize Weights
        weight = 2 * np.random.random((X.shape[1], 1)) - 1

        for i in range(self.niter):
            # Weighted sum of inputs / weights
            weighted_sum = np.dot(X, weight)

            # Activate!
            activated_output = self.sigmoid(weighted_sum)

            # Calculate absolute error
            error = y - activated_output

            adjustments = error * sigmoid_derivative(weighted_sum)

            # Update the Weights
            weight += np.dot(X.T, adjustments)

#     def net_input(self, X):
#         return np.dot(self.weight, X)

    def predict(self, X):
        return self.sigmoid(self.net_input(X.T))
```
