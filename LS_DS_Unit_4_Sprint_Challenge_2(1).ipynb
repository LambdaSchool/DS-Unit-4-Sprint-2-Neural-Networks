## 1. Define the following terms:

- **Neuron:** A neuron is a mathematical construct that takes in some number of inputs, performs mathematical operations on them, and then multiplies them by the output of an activation function. It then returns this result as its output.
- **Input Layer:** The input layer is the first layer of a neural network. It is the only layer that interacts directly with the inputs.
- **Hidden Layer:** A hidden layer of a neural network is an intermediate layer, which is not the input layer or the output layer.
- **Output Layer:** The output layer of a neural network is the final layer, which produces the final outputs.
- **Activation:** The activation function is a function that determines whether or not the neuron produces an output, and how strong that output is.
- **Backpropagation:** Backpropagation is the standard method by which a neural network learns. Weights of neurons are adjusted in reverse order, starting with the output layer, moving through the hidden layers last to first, and finally reaching the input layer.

Once you've trained your model, report your accuracy. You will not be able to achieve more than ~50% with the simple perceptron.
```python
# Start your candy perceptron here

X = candy[['chocolate', 'gummy']].values
y = candy['ate'].values
# # X = X.T
X.shape[0]
```

```python
class Perceptron(object):
    def __init__(self, rate = 0.01, iter = 100):
        self.rate = rate
        self.iter = iter
    
    def fit(self, X, y):
        self.weight = np.zeros(1 + X.shape[1])
        self.errors = []
        for i in range(self.iter):
            err = 0
            for xi, target in zip(X, y):
                delta_w = self.rate * (target - self.predict(xi))
                self.weight[1:] += delta_w * xi
                self.weight[0] += delta_w
                err += int(delta_w != 0.0)
            self.errors.append(err)
        return self
    
    def net_input(self, X):
        return np.dot(X, self.weight[1:]) + self.weight[0]
    
    def predict(self, X):
        return np.where(self.net_input(X) >= 0.0, 1, -1)
    
    def accuracy(self, X):
        return 1 - (self.errors[-1] / X.shape[0])
```

```python
nn = Perceptron()
```

```python
nn.fit(X, y)
```

```python
nn.accuracy(X)
```

Final accuracy is pretty terrible, at about 37%. This is because this dataset is essentially an XOR gate, which the simple perceptron cannot deal with. The output of an XOR gate is not linearly separable.

```python
class NeuralNetworks:
    def __init__(self):
        self.inputs = 2
        self.hiddenNodes = 5
        self.outputNodes = 1
        self.weights1 = np.random.rand(self.inputs, self.hiddenNodes)
#         weights1 is 2, 5
        self.weights2 = np.random.rand(self.hiddenNodes, self.outputNodes)
#        weights2 = 5,1
    
    def sigmoid(self, s):
        return 1 / (1+np.exp(-s))
    
    def sigmoidPrime(self, s):
        return s * (1 - s)
    
    def feed_forward(self, X):
        self.hidden_sum = np.dot(X, self.weights1)
#         10000, 2 * 2, 5 = 10000, 5
        self.activated_hidden = self.sigmoid(self.hidden_sum)
#        shape unaffected                                      
        self.output_sum = np.dot(self.activated_hidden, self.weights2)
#         10000, 2 * 5, 1 = 10000, 1                                     
        self.activated_output = self.sigmoid(self.output_sum)
#        shape unaffected                                      
        return self.activated_output
    
    def backward(self, X,y,o):
        self.o_error = y - o
#        10000, 1 - 10000, 1 = shape unaffected                                      
        self.o_delta = (self.o_error * self.sigmoidPrime(o))
#        10000, 1 * 10000, 1 = 10000, 1                                     
        self.z2_error = self.o_delta.dot(self.weights2.T)
#        10000, 1 * 1, 5 = 10000, 5                                      
        self.z2_delta = self.z2_error * self.sigmoidPrime(self.activated_hidden)
#        10000, 5 * 10000, 5 = 10000, 5                                      
        self.weights1 += X.T.dot(self.z2_delta)
#        2, 10000 * 10000, 5 = 2, 5                                 
        self.weights2 += self.activated_hidden.T.dot(self.o_delta)
#         5, 10000 * 10000, 1 = 5, 1
    def train(self, X, y):
        o = self.feed_forward(X)
        self.backward(X, y, o)
    def accuracy(self):
        return 1 - np.mean(abs(self.o_error))
```

```python
nnn = NeuralNetworks()
```

```python
for _ in range(10):
    nnn.train(X, y)
```

```python
nnn.accuracy()
```

This more complex neural network is able to beat random chance, but not by much. The complexity added by hidden layers allows for classification beyond the simple linear separation possible with a basic perceptron. I will have to take your word on the chocolate gummy bears.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
df = pd.read_csv('https://raw.githubusercontent.com/ryanleeallred/datasets/master/heart.csv')
df = df.sample(frac=1)
print(df.shape)
df.head()
```

```python
X = df.drop(['target'], axis=1)
y = df.target
X = X.to_numpy()
y = y.to_numpy()
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
# Without scaling the model gets stuck at a local minimum
```

```python
def create_model():
    model = Sequential([
    Dense(13, input_shape=(13,)),
    Activation('relu'),
    Dense(13),
    Activation('relu'),
    Dense(13),
    Activation('relu'),
    Dense(1),
    Activation('sigmoid')
    ])
    model.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"], use_multiprocessing=True)
    return model
model = KerasClassifier(build_fn=create_model, verbose=1)
```

```python
epochs = 100
batch_size = 100
model.fit(X, y, epochs=epochs, batch_size=batch_size, verbose=1, validation_split = 0.25)
```

```python
param_grid = {'batch_size': [10, 20, 40, 60],
              'epochs': [100, 120, 140, 160, 180, 200]}

rando = GridSearchCV(estimator=model, scoring = 'accuracy', param_grid=param_grid, n_jobs=-1)
random_result = rando.fit(X, y)
print(random_result.best_score_, random_result.best_params_)
```
