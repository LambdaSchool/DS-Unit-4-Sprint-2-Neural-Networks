## Write your own perceptron code that can correctly classify (99.0% accuracy) a NAND gate. 

| x1 | x2 | y |
|----|----|---|
| 0  | 0  | 1 |
| 1  | 0  | 1 |
| 0  | 1  | 1 |
| 1  | 1  | 0 |

```python
import pandas as pd
import numpy as np
data = { 'x1': [0,1,0,1],
         'x2': [0,0,1,1],
         'y':  [1,1,1,0]
       }

df = pd.DataFrame.from_dict(data).astype('int')
df.head()
```

```python
##### Your Code Here #####
list(df)
```

```python
def sigmoid(x):
    return 1 / (1+ np.exp(-x))

def sigmoid_derivative(x):
    sx = sigmoid(x)
    return sx * (1 - sx)
```

```python
inputs = np.array(df[["x1","x2"]])

inputs = np.append(inputs, [[1],[1],[1],[1]],axis = 1)

correct_outputs = np.array(df[["y"]])
```

```python
inputs
```

```python
correct_outputs
```

```python
# Let's select random weights to start.
weights = (2 * np.random.random((3,1)) - 1)
weights
```

```python
weighted_sum = np.dot(inputs, weights)
weighted_sum
```

```python
# These are the predicted outputs.
activated_outputs = sigmoid(weighted_sum)
activated_outputs
```

```python
correct_outputs
```

```python
# This is the measuring error.
error = correct_outputs - activated_outputs
error
```

```python
# Let's adjust the weight.
adjustments = error * sigmoid_derivative(weighted_sum)
adjustments
```

```python
# These are the new weights.
weights += np.dot(inputs.T, adjustments)
weights
```

```python
# Updated weights over 1000 iterations.
for iteration in range(1000):
    weighted_sum = np.dot(inputs, weights)
    activated_output = sigmoid(weighted_sum)
    error = correct_outputs - activated_output
    adjustments = error * sigmoid_derivative(weighted_sum)
    weights += np.dot(inputs.T, adjustments)
    
print("Weights after training")
print(weights)

print("Output after training")
print(activated_output)    
```

```python
from sklearn.preprocessing import MinMaxScaler, Normalizer

target = ['Outcome']
feats = list(diabetes)[:-1]
X = diabetes[feats]
y = np.array(diabetes[target])

scaler = MinMaxScaler()
X = scaler.fit_transform(X)
X
```

```python
##### Update this Class #####

class Perceptron:
    
    def __init__(self, niter = 10):
        self.niter = niter
    
    def __sigmoid(self, x):
        return 1 / (1+ np.exp(-x))
    
    def __sigmoid_derivative(self, x):
        sx = sigmoid(x)
        return sx * (1 - sx)   

    def fit(self, X, y):
        """Fit training data
        X : Training vectors, X.shape : [#samples, #features]
        y : Target values, y.shape : [#samples]
        """
        # Add bias var
        bias = []
        for i in range(len(X)):
            bias += [[1]]
        X = np.append(X, bias,axis = 1)
        
        # Randomly Initialize Weights
        self.weights = (2 * np.random.random((X.shape[1], 1)) - 1)
        
        for i in range(self.niter):
            # Weighted sum of inputs / weights
            weighted_sum = np.dot(X, self.weights)
            # Activate!
            activated_output = sigmoid_derivative(weighted_sum)
            # Cac error
            error = y - activated_output
            # Update the Weights
            adjustments = error * sigmoid_derivative(weighted_sum)
            self.weights += np.dot(X.T, adjustments)
        return self 

    def activate(self, X):
        """Calculate net input"""
        # Add bias var
        bias = []
        for i in range(len(X)):
            bias += [[1]]
        X = np.append(X, bias,axis = 1)
        
        weighted_sum = np.dot(X, self.weights)
        activated_output = sigmoid(weighted_sum)
        return activated_output
    
    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.activate(X) >= .5, 1, 0)
```

```python
pn = Perceptron()
```

```python
pn.fit(X, y)
```
